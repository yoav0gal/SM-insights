```python
import string

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
```

```python
from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
from collections import Counter

def word_counts(df):
    cnt = Counter()
    for text in df["text"].values:
        for word in text.split():
            cnt[word] += 1
            
    return cnt
```

```python
def remove_freqwords(text, cnt):
    FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])
```

```python
def remove_rarewords(text, cnt):
    n_rare_words = 10
    RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])
```

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])
```

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
import inflect

p = inflect.engine()

def convert_numbers_to_words(text):
    words = text.split()
    converted_words = [p.number_to_words(word) if word.isdigit() else word for word in words]
    return " ".join(converted_words)
```

```python
def remove_numbers(text):
    return re.sub(r'\d+', '', text)
```

```python
chat_words_map_dict = {}
chat_words_list = []
for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded
chat_words_list = set(chat_words_list)

def chat_words_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words_list:
            new_text.append(chat_words_map_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

## Format the datasets using the data prepreocessors

```python
def process_df(df, text_column, new_column):
    df[new_column] = df[text_column]\
        .apply(lambda x: x.lower()) \
        .apply(lambda x: remove_urls(x)) \
        .apply(lambda x: remove_html(x)) \
        .apply(lambda x: convert_emojis(x)) \
        .apply(lambda x: convert_emoticons(x)) \
        .apply(lambda x: remove_emoji(x)) \
        .apply(lambda x: remove_numbers(x)) \
    
    df = df[df[new_column].str.split().str.len() > 3]
    df = df[df[new_column].map(lambda x: x.isascii())]
    return df
```

```python
import google.generativeai as genai
from bertopic import BERTopic
import os
from tenacity import retry, wait_exponential, stop_after_attempt, retry_if_exception_type
from google.api_core.exceptions import ResourceExhausted
import time

# 1. Initialize Gemini client

genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

@retry(
    retry=retry_if_exception_type(ResourceExhausted),
    wait=wait_exponential(multiplier=1, min=1, max=60),
    stop=stop_after_attempt(5),
)
def embed_batch(docs):
    return genai.embed_content(
        model="models/text-embedding-005",
        content=docs).embeddings
    
def gemini_embed(documents, batch_size=50, rpm_limit=1400):
    embeddings, batch = [], []
    interval = 60 / rpm_limit * batch_size
    for idx, doc in enumerate(documents, 1):
        batch.append(doc)
        if len(batch) == batch_size or idx == len(documents):
            embs = embed_batch(batch)
            embeddings.extend(embs)
            batch.clear()
            time.sleep(interval)  # throttle to stay under RPM
    return embeddings
```
